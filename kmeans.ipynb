```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans

# Generate random 2D data
np.random.seed(42)
X = np.random.rand(300, 2)  # 300 points in 2D space

# Store cost function values over multiple runs
cost_values = []

# Run K-Means for different numbers of clusters and store the cost function (inertia)
max_clusters = 10  # Trying different K values from 1 to 10
for k in range(1, max_clusters + 1):
    kmeans = KMeans(n_clusters=k, max_iter=50, random_state=42, n_init=10)
    kmeans.fit(X)
    cost_values.append(kmeans.inertia_)  # Store final cost (inertia)

# Plot cost function vs. number of clusters (Elbow Method)
plt.figure(figsize=(8, 5))
plt.plot(range(1, max_clusters + 1), cost_values, marker='o', linestyle='-')
plt.xlabel("Number of Clusters (K)")
plt.ylabel("Cost Function (Inertia)")
plt.title("Elbow Method: Choosing the Best K")
plt.grid(True)
plt.show()
```
