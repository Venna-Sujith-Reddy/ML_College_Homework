```python
import numpy as np
import matplotlib.pyplot as plt

# Define weight vectors for three classes (C1, C2, C3)
w1 = np.array([1, 1])
w2 = np.array([-1, 1])
w3 = np.array([0, -1])
b1, b2, b3 = -0.5, 0.5, -0.2  # Bias terms

# Create a grid of points
x_range = np.linspace(-2, 2, 100)
y_range = np.linspace(-2, 2, 100)
X, Y = np.meshgrid(x_range, y_range)
points = np.c_[X.ravel(), Y.ravel()]

# Compute discriminant functions
y1 = np.dot(points, w1) + b1
y2 = np.dot(points, w2) + b2
y3 = np.dot(points, w3) + b3

# Assign class based on the highest discriminant function value
class_labels = np.argmax(np.c_[y1, y2, y3], axis=1).reshape(X.shape)

# Plot decision regions
fig, ax = plt.subplots(figsize=(6,6))
ax.contourf(X, Y, class_labels, alpha=0.3, levels=2, cmap="coolwarm")

# Plot decision boundaries (where discriminant functions are equal)
ax.contour(X, Y, y1 - y2, levels=[0], colors='r', linewidths=2, linestyles='dashed', label="Boundary C1 vs. C2")
ax.contour(X, Y, y2 - y3, levels=[0], colors='g', linewidths=2, linestyles='dashed', label="Boundary C2 vs. C3")
ax.contour(X, Y, y3 - y1, levels=[0], colors='b', linewidths=2, linestyles='dashed', label="Boundary C3 vs. C1")

# Plot weight vectors
ax.quiver(0, 0, w1[0], w1[1], color='r', angles='xy', scale_units='xy', scale=1, label="w1")
ax.quiver(0, 0, w2[0], w2[1], color='g', angles='xy', scale_units='xy', scale=1, label="w2")
ax.quiver(0, 0, w3[0], w3[1], color='b', angles='xy', scale_units='xy', scale=1, label="w3")

# Labels and title
ax.set_xlabel("x1")
ax.set_ylabel("x2")
ax.set_title("Classification Using Discriminant Functions")
ax.legend()
ax.grid(True, linestyle='--', linewidth=0.5)

# Show the plot
plt.show()
```
