```python
import numpy as np

import matplotlib.pyplot as plt
```

```python
X = np.array([[0.5, 1.5], [1,1], [1.5, 0.5], [3, 0.5], [2, 2], [1, 2.5]])
y = np.array([0, 0, 0, 1, 1, 1]).reshape(-1,1) 
```

```python
fig,axis=plt.subplots(1,1,figsize=(10,10))
axis.axis([0,4,0,3.5])

axis.set_xlabel('$x_1$')
axis.set_ylabel('$x_2$')

c1=X[y.flatten()==0]
c2=X[y.flatten()==1]
axis.scatter(c1[:,0],c1[:,1],marker="X",c="red")
axis.scatter(c2[:,0],c2[:,1],marker="X",c="blue")
plt.show()
```
