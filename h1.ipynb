```python
w = 100
b = 100
print(f"w: {w}")
print(f"b: {b}")
```

```python
def compute_model(x,w,b):
    m = x.shape[0]
    print(m)
    f_wb = np.zeros(m)
    for i in range(m):
        f_wb[i] = w * x[i] + b
        
    return f_wb
```

```python
tmp_f_wb = compute_model(X_train, w, b)
```

```python
plt.plot(X_train,tmp_f_wb,c='b',label="Our Predictions")
plt.scatter(X_train,Y_train,marker="x",c="r")
plt.legend()
plt.show()
```
