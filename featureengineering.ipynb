```python
def gradient(w,b,x,y):
    m=x.shape[0]
    dj_dw=0
    dj_db=0
    for i in range(m):
        dj_dw+=(w*x[i,0]+b-y[i])*x[i,0]
        dj_db+=(w*x[i,0]+b-y[i])
    return dj_dw/m,dj_db/m
```
